```python
import re
from nltk.corpus import stopwords

# Initialize Greek stopwords
stop_words = set(stopwords.words('greek'))

# Function to load suffixes from a text file
def load_suffixes(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        return [line.strip() for line in f if line.strip()]

# Function to load tones from a text file
def load_tones(file_path):
    tones = {}
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            original, normalized = line.strip().split(',')
            tones[original] = normalized
    return tones

# Load suffixes and tones from files
suffixes = load_suffixes(r'C:\Users\..........\suffixes.txt')
tones = load_tones(r'C:\Users\..........\tones.txt')

def remove_tones(text):
    for accented, unaccented in tones.items():
        text = text.replace(accented, unaccented)
    return text

def remove_stopwords(text):
    return ' '.join([word for word in text.lower().split() if word not in stop_words])

def stem(word, suffixes):
    done = len(word) <= 3
    if not done:
        for suffix in suffixes:
            if stemma(word, suffix):
                word = word[:len(word) - len(suffix)]
    return word

def stemma(word, suffix):
    return word[len(word) - len(suffix):] == suffix

def lemmatize_words(text, suffixes):
    return ' '.join([stem(word, suffixes) for word in text.split()])

def clean_text(text, suffixes):
    if text is None:
        return ''
    
    text = remove_tones(text)
    text = remove_stopwords(text)
    text = text.upper()
    
    # Αφαίρεση URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # Αφαίρεση @mentions
    text = re.sub(r'@\w+', '', text)

    # Αφαίρεση hashtags
    text = re.sub(r'#\w+', '', text)

    # Αφαίρεση σημεία στίξης εκτός από κενά
    text = re.sub(r'[^\w\sΑ-Ω]', '', text)
    
    # Remove non-Greek words
    text = re.sub(r'\b[^\sΑ-Ω]+\b', '', text)

    # Αφαίρεση περιττών κενών
    text = re.sub(r'\s+', ' ', text).strip()
    
    # Lemmatization (using stemming here)
    text = lemmatize_words(text, suffixes)

    return text

# Διαδρομή αρχείου
input_file = r'C:\Users\...........\DoW.txt'
output_file = r'C:\Users\............\DoW_2.txt'

# Διαβάζουμε το αρχείο εισόδου
with open(input_file, 'r', encoding='utf-8') as file:
    words = file.read().splitlines()

# Καθαρισμός κειμένου και αφαίρεση duplicates
cleaned_words = {clean_text(word, suffixes) for word in words}  # Χρησιμοποιούμε set για να αφαιρέσουμε τα duplicates

# Ταξινόμηση αλφαβητικά
sorted_words = sorted(cleaned_words)

# Αποθήκευση στο νέο αρχείο
with open(output_file, 'w', encoding='utf-8') as file:
    for word in sorted_words:
        file.write(word + '\n')
```

```python
import re
from nltk.corpus import stopwords
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Initialize Greek stopwords
stop_words = set(stopwords.words('greek'))

# Function to load suffixes from a text file
def load_suffixes(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        return [line.strip() for line in f if line.strip()]

# Function to load tones from a text file
def load_tones(file_path):
    tones = {}
    with open(file_path, 'r', encoding='utf-8') as f:
        for line in f:
            original, normalized = line.strip().split(',')
            tones[original] = normalized
    return tones

# Load suffixes and tones from files
suffixes = load_suffixes(r'C:\Users\............\suffixes.txt')
tones = load_tones(r'C:\Users\.........\tones.txt')

def remove_tones(text):
    for accented, unaccented in tones.items():
        text = text.replace(accented, unaccented)
    return text

def remove_stopwords(text):
    return ' '.join([word for word in text.lower().split() if word not in stop_words])

def clean_text(text, suffixes):
    if text is None:
        return ''
    
    text = remove_tones(text)
    text = remove_stopwords(text)
    text = text.upper()
    
    # Αφαίρεση URLs
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

    # Αφαίρεση @mentions
    text = re.sub(r'@\w+', '', text)

    # Αφαίρεση hashtags
    text = re.sub(r'#\w+', '', text)

    # Αφαίρεση σημεία στίξης εκτός από κενά
    text = re.sub(r'[^\w\sΑ-Ω]', '', text)
    
    # Αφαίρεση περιττών κενών
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# Διαδρομή αρχείου
input_file = r'C:\Users\..........\DoW.txt'

# Διαβάζουμε το αρχείο εισόδου
with open(input_file, 'r', encoding='utf-8') as file:
    words = file.read().splitlines()

# Καθαρισμός κειμένου και συγχώνευση όλων των λέξεων
cleaned_texts = [clean_text(word, suffixes) for word in words]
all_cleaned_text = ' '.join(cleaned_texts)

# Δημιουργία Bag of Words (Counter)
bow = Counter(all_cleaned_text.split())

# Δημιουργία σύννεφου λέξεων
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(bow)

# Εμφάνιση του σύννεφου λέξεων
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Αφαίρεση άξονα
plt.show()

# Εμφάνιση της συχνότητας κάθε λέξης ταξινομημένα
print("Συχνότητα εμφάνισης κάθε λέξης (ταξινομημένα):")
for word, freq in bow.most_common():
    print(f"{word}: {freq}")
```
