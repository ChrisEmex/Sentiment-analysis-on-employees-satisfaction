```python
import pandas as pd

# Φόρτωση του αρχείου Excel
file_path = r"C:\Users\..........\answers_Yes.xlsx"
df = pd.read_excel(file_path, sheet_name='Sheet1')  # Προσαρμόστε το sheet_name αν χρειάζεται

# Έλεγχος για τη στήλη "Q7"
if 'Q7' in df.columns:
    # Απαντήσεις προς καταμέτρηση
    options = [
        "Αύξηση μισθού",
        "Καλύτερο επικοινωνιακό περιβάλλον με τους συναδέλφους",
        "Καλύτερη επικοινωνία με τα ανωτέρω κλιμάκια",
        "Καλύτερο περιβάλλον εργασίας (εγκαταστάσεις, υποδομές, υλικοτεχνικό υλικό)",
        "Όλα λειτουργούν όπως πρέπει"
    ]
    
    # Λεξικό για την αποθήκευση των μετρήσεων
    counts = {option: 0 for option in options}
    
    # Καταμέτρηση της εμφάνισης κάθε απάντησης
    for response in df['Q7'].dropna():  # Παράλειψη κενών γραμμών
        for option in options:
            if option in response:
                counts[option] += 1

    # Εκτύπωση των αποτελεσμάτων
    for option, count in counts.items():
        print(f"{option}: {count}")
else:
    print("Η στήλη 'Q7' δεν βρέθηκε στο αρχείο.")
```

```python
import matplotlib.pyplot as plt

# Data for the bar plot
responses = {
    "Salary increase": 153,
    "Better communication environment with colleagues": 56,
    "Better communication with senior management": 95,
    "Better working environment (facilities, infrastructure, equipment)": 93,
    "Everything is functioning as it should": 14
}

# Sort responses by value (from smallest to largest)
sorted_responses = dict(sorted(responses.items(), key=lambda item: item[1]))

# Plotting
plt.figure(figsize=(6, 9))
plt.bar(sorted_responses.keys(), sorted_responses.values(), color='skyblue')
plt.title('Distribution of Responses in Question Q7')

plt.ylabel('Total Count')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()

plt.show()
```
